'NN' : 명사, 단수형,    
'CC' : 접속사    
'VBN' : 동사 과거   
'VBZ' : 동사 현재 3인칭  
'DT' : 한정사(명사 앞에 붙는 the, some, my 등과 같은 말들)  
'IN' : 접속사 전치사   
'JJ' : 형용사 서수 (first, second..)  
'RB' : 부사    
'VBD' : 과거 동사   
'NNS' : 명사 복수형  
'PRP' : 재귀명사,3인칭  
'VB' : 동사 기본형  
'MD' : 조동사  
'VBG' : 동사 현재진행형  
'VBP' : 동사 타동사  
'WDT' : 복합관계대명사  
'RP' : 부사(특히 동사 뒤에 쓰여 위치·동작의 방향 등을 나타내는 부사)  
'WRB' : 부사   
'TO' : to  
'CD' : 기수   
'WP$' : whose  
'PRP$' : 소유 대명사  
'RBS' : 부사 최상급  
'PDT' : 전치 한정사  
'WP' : wh-대명사  
'FW' : 외래어  
'JJS' : 형용사 최상급  
'JJR' : 형용사 비교급   
'RBR' : 부사 비교급   
'EX' : there  
'NNP' : 고유명사
