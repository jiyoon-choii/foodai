```python
from selenium import webdriver
import urllib, requests
import time
from base64 import b64encode # byte배열을 base64로 변경함.
# 모든 페이지
url_list=list()
driver = webdriver.Chrome('C:/Users/HANYEONGJAE/Desktop/cr/chromedriver.exe')
for h in range(0,1020,30):
    print("#"*50)
    print(h , "|", "url start")
    print("="*50)
    print(": url 들고오기 시작")
    if h == 0:
        url='https://www.tripadvisor.com/Search?q=south%20korea&searchSessionId=FEA1095D400E57A72FD10023F998B9171588658590823ssid&searchNearby=false&sid=BB26669DA95EE28FEEAE860A2E877D901588658599003&blockRedirect=true&rf=3&ssrc=e'
    else:
        url='https://www.tripadvisor.com/Search?q=south%20korea&searchSessionId=FEA1095D400E57A72FD10023F998B9171588658590823ssid&searchNearby=false&sid=BB26669DA95EE28FEEAE860A2E877D901588658599003&blockRedirect=true&rf=5&ssrc=e&o={}'.format(h)
    print(h, ": url 들고오기 끝남")
    
    try:
        driver.get(url)
        time.sleep(3)
    except:
        time.sleep(300)
        driver.get(url)
    
    # a 태그 안에 있으니까 a tag를 찾는다
    print(h, ": a 태그 가져오기 시작")
    li= driver.find_elements_by_tag_name("a")
    print(h, ": a 태그 가져오기 끝남")
    
    # href 가져오기
    print(h, ": href 가져오기 시작")
    k=list()
    for i in li:
        k.append(i.get_attribute("href"))
    print(h, ": href 가져오기 끝남")

    # None으로 되어있는 부분 빼고 추출
    print(h, ": None 빼기 시작")
    li=list()
    for i in k:
        if type(i)!= type(None):
            li.append(i)
    print(h, ": None 빼기 끝남")
    
    # Review href 만 추출
    print(h, ": href 추출 시작")
    
    for s in range(1,len(li),1):
        #print(s)
        if "#REVIEWS" in li[s]:
            #print('True')
            url_list.append(li[s])
    print(h, ": href 추출 끝남")
    print("="*50)
    
    if len(li) == 0 :
        break
    
    print(h , "|", "url end")
```
