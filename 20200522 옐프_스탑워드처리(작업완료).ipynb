- 1. df['review']의 값(문장)을 단어 단위로 나누기 (대문자->소문자 처리)

```python
tag_count = []
tags = []

# review 의 상위 row 10000개 가져오기  
for review in df['review'][:5000] :
    # 대문자 -> 소문자로 바꾸기
    review = review.lower()
#     print(review)
#     print(type(review))

#     nltk에 내장된 stopword 먼저 제거 - 시도했으나 사용 x
#     stop_words = set(stopwords.words('english')) 
#     print(stop_words)
#     if stop_words in review :
#         pass
#         review = review.replace('',review.lower())
     
#     review = review[i] 
    word_tokens = word_tokenize(review)
#     if word_tokens in stop_words :
#     # review 가 문장이기 때문에, 
#     # 문장을 단어 단위로 나눠줌
    count = Counter(word_tokens)
#         # 각 단어가 몇 개인지 count 에 담기     
    for n,c in count.most_common()[:7000] :
#             # count.mos_common(n) : 가장 빈도 수가 높은 단어 n 개를 뽑는다 
#             # n : 단어 / c : 빈도
#     #         print(c)
        dics = dict()
        dics = {'tag':n, 'count':c}
        tag_count.append(dics)
#             # [{단어,빈도},{},....] 를 tag_count 리스트 에 담아준다
        tags.append(dics['tag'])
#                 # [dics의 value(c),c,c,......] 를 tags 리스트에 담아준다 
```
