```python
def read_txt(filename,sep='') :
    str = ''
    
    file = open(filename,'r',encoding='utf-8')
    
    str = file.readlines()
    
    for i in range(0,len(str)) :
        str[i] = str[i].strip().split(sep)
    return(str)

    file.close()
    
all_add = read_txt('./data/value.txt',',')
```
