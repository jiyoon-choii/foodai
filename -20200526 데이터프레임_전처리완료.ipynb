- 함수를 써보자 ! 

```python
def wordChg( data ):
    #print( all_data4['menu'][idx])
    tmp = list()
    #for word in all_data4['menu'][idx].replace('[','').replace(']','').split(","):
    for word in data.replace('[','').replace(']','').split(","):        
        tmp.append(word.strip().replace("'",''))
    return tmp
wordChg("[]")
```
